```python
id = []
casts = []
genres = []
title = []
vote_count = []
ratings = []
for i in range(len(contents)):
    id.append(contents[i]['fields']['id'])
    casts.append(contents[i]['fields']['casts'])
    title.append(contents[i]['fields']['title'])
    vote_count.append(contents[i]['fields']['vote_count'])
    ratings.append(contents[i]['fields']['vote_average'])
    # genres.append(contents[i]['fields']['genres'])
    temp = []
    for j in range(len(contents[i]['fields']['genres'])):
        tmp_name = genres_dict[contents[i]['fields']['genres'][j]]
        temp.append(tmp_name)
        # genres.append(genres_data[genres_data['pk'] == contents[i]['fields']['genres'][j]]['fields'])
    genres.append(temp)
    
```

```python
movies_rating = pd.DataFrame({'id': id,
                              'casts': casts,
                              'genres': genres,
                              'title': title,
                              'vote_count': vote_count,
                              'ratings': ratings})
```

```python
movies_rating['genres']=movies_rating['genres'].apply(lambda x : [ y['name'] for y in x])
movies_rating[['genres']][:1]
```

```python
movies_rating['genres_literal']=movies_rating['genres'].apply(lambda x : (' ').join(x))
```

```python
count_vect=CountVectorizer(min_df=0, ngram_range=(1, 2))
genre_mat=count_vect.fit_transform(movies_rating['genres_literal'])
print(genre_mat.shape)
```

```python
genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
```

```python
genre_sim_sorted_ind=genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:1])
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = movies_rating[movies_rating['title'] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]

    similar_indexes = similar_indexes.reshape(-1)

    return df.iloc[similar_indexes]
```

```python
similar_movies = find_sim_movie(movies_rating, genre_sim_sorted_ind, '더 배트맨', 10)
similar_movies[['title', 'ratings']]
```

```python
percentile = 0.6
m = movies_rating['vote_count'].quantile(percentile)
C = movies_rating['ratings'].mean()

def weighted_vote_avg(record):
    v = record['vote_count']
    R = record['ratings']

    return ( (v/(v+m) * R) + (m/(m+v)) * C)

movies_rating['weighted_vote'] =movies_rating.apply(weighted_vote_avg, axis = 1)
```

```python
movies_rating[['title', 'vote_count', 'ratings', 'weighted_vote']].sort_values('weighted_vote', 
                                                                               ascending=False)[:10]
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    similar_indexes = sorted_ind[title_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)
    similar_indexes = similar_indexes[similar_indexes != title_index]

    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]
```

```python
similar_movies = find_sim_movie(movies_rating, genre_sim_sorted_ind, '배트맨 비긴즈', 15)
similar_movies[['title', 'ratings', 'weighted_vote']]
```
